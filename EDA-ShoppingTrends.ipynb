```python
data['Season'].value_counts().plot()
plt.xlabel('Season')
plt.ylabel('Count')
plt.title('Count of Purchases in Each Season')
plt.show()
```

***Spring** has the highest purchase count by a significant margin, followed by **Fall** and **Winter**.*



```python
data['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Purchases by Category')
plt.show()
```

*The pie chart shows that **Clothing** and **Accessories** dominate the purchase distribution, with these categories holding the largest shares.*


```python
data.boxplot(column= 'Review Rating', by='Gender')
plt.xlabel('Gender')
plt.ylabel('Review Rating')
plt.title('Review Rating Distribution by Gender')
plt.show()
```

*The **Male** review ratings have a **Larger range** and a **Higher mean rating** compared to the **Female** ones.*

```python
data['Purchase Amount (USD)'].plot(kind='hist', bins = 10)
plt.xlabel('USD')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Amount Distribution')
plt.show()
```

*Customers are more likely to buy products with **low prices ($30)** and **high prices ($100)**.*

```python
data.groupby('Color')['Review Rating'].mean().head(5).plot(kind = 'bar')
plt.xlabel('Color')
plt.ylabel('Mean Review Rating')
plt.title('Mean Review Rating for Each Color')
plt.show()
```

*These are the **Top 5 color** ratings. The most popular are **Dark** and **Formal** colors. This correlation is logical, as these colors align with the preferences of predominantly middle-aged male shoppers.*


```python
data.groupby('Payment Method')['Purchase Amount (USD)'].sum().plot(kind='bar')
plt.title('Sum of Purchase Amount by Payment Method')
plt.ylabel('')
plt.show()
```

*The bar chart shows that **Credit Card** is the top payment method by total purchase amount.*


```python
data.boxplot(column='Purchase Amount (USD)', by = 'Frequency of Purchases')
plt.xlabel('Frequency of Purchases') 
plt.ylabel('Purchase Amount (USD)')
plt.title('Purchase Amount Distribution by Frequency of Purchases')
plt.show()
```

*The **bi-weekly** purchase frequency has the highest median purchase amount, indicating that customers who shop every two weeks tend to spend more per purchase compared to other frequencies. This suggests that bi-weekly shoppers may be making larger or more frequent purchases.*


```python
data.groupby('Season')['Purchase Amount (USD)'].sum().plot()
plt.xlabel('Season')
plt.ylabel('Sum of Purchase Amount (USD)')
plt.title('Sum of Purchase Amount by Season')
plt.show()
```

*While **Spring** has the highest purchase count, **Fall** and **Winter** lead in total purchase amounts.*
